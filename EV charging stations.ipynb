# Scrapping locations of EV charging stations in China

This project was intended for an [article](https://www.scmp.com/business/companies/article/3105843/do-electric-cars-have-range-go-distance-chinas-rural-towns-and) in an EV series by the South China Morning Post. I was one of the reporters participate in the story. 

I wanted to show the distribution of all the EV charging stations in China by mapping them, and see if the uneven distribution would be a bottle neck for the country's push of the EV to "go rural" - go into the rural markets. I scraped this [website](http://admin.bjev520.com/jsp/beiqi/pcmap/do/index.jsp) of Beiqi New EV.

But my knowledge about the front-end visualization is very limited. Though I got the all the data, I didn't manage to visualize them (and the infographics team was too busy to take care of this too at that time), so this idea was not realized in the final story, unfortunaly.

```python
import requests
def getlnglat(address):
    url = 'http://api.map.baidu.com/geocoding/v3/'
    output = 'json'
    ak = '269klMZwcP6BTEqubRhnFNFgZcTUwb4c'
   
    #生成完整url
    uri = url + '?' + 'address=' + address  + '&output=' + output + '&ak=' + ak+ "&callback=showLocation //GET请求
    html = requests.get(uri)#获取网页内容
    bs_getDetail = BeautifulSoup(html.text,'lxml')  #解析网页  
    #找到存放经纬度数据的文字，转换为字典，取得相应经纬度
    lat = eval(bs_getDetail.p.text)["result"]["location"]["lat"]
    lng = eval(bs_getDetail.p.text)["result"]["location"]["lng"]
   
    lng_lat = [lng, lat]
    return lng_lat
```

```python
def BaiduQuery(address, currentkey):
    """
    address convert lat and lng
    :param address: address
    :param currentkey: AK
    :return: places_ll
    """
    url = 'http://api.map.baidu.com/geocoding/v3/?'
    params = {
        "address": address,
        "output": 'json',
        "ak": currentkey,
    }
    response = requests.get(url, params=params)
    answer = response.json()
    if answer['status'] == 0:
        tmpList = answer['result']
        coordString = tmpList['location']
        coordList = [coordString['lng'], coordString['lat']]
        places_ll.append([address, float(coordList[0]), float(coordList[1])])
        print([address, float(coordList[0]), float(coordList[1])])
    else:
        return -1
```
